```python
import matplotlib.pyplot as plt
import numpy as np
# Define a function to calculate the electric field at a given distance from a source
def electric_field(distance, voltage):
      field = voltage / (4 * np.pi * distance**2)
      return field
# Define the voltage and distance for the source
voltage = 10
distance = 5
# Calculate the electric field at the given distance
field = electric_field(distance, voltage)
# Plot the electric field lines
plt.plot(distance, field)
plt.xlabel('Distance (m)')
plt.ylabel('Electric field (V/m)')
plt.title('Inverse square law')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Constants
G = 6.6743e-11    # Gravitational constant
M = 7.34e22    # Mass of the earth (in kg)
R = 6371e3    # Radius of the earth (in m)
# Initial conditions
r = 1e3    # Initial distance of the moon (in m)
v = 2e3    # Initial velocity of the moon (in m/s)
t = 0    # Time (in s)
# Time loop
for t in range(0, 365*24*3600, 1):
      r_new = r**2 * (G*M/(4*R**2))    # Inverse square law
      v_new = v + r_new/r * (G*M/(4*R**2))    # Acceleration
      r = r_new
      v = v_new
      plt.plot(t, r, 'o-')
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.show()
```
